```python
import pickle
import navigator
import pandas as pd

pd.set_option('display.max_rows', 500)

def read_routes(org, des):
    filename = f'{org}-{des}.pkl'

    with open(filename, 'rb') as f:
        data = pickle.load(f)

    org_routes, des_routes, bikes, bike_routes = data['org_routes'], data['des_routes'], data['bikes'], data['bike_routes']

    rows = []
    for i, srt in enumerate(bikes):
        (_, (srt_name, srt_lat, srt_lng, _, _, srt_coord)) = srt
        for j in range(i+1, len(bikes)):
            try:
                (_, (end_name, end_lat, end_lng, _, _, end_coord)) = bikes[j]
                route_id = f'{i}-{j}'
                srt_time = navigator.calc_timedelta(org_routes[j][0]['time'])
                bike_time = navigator.calc_timedelta(bike_routes[route_id][0]['time'])
                end_time = navigator.calc_timedelta(des_routes[j][0]['time'])
                total_time = srt_time + bike_time + end_time
                rows.append([route_id, org, str(srt_time), srt_name, str(bike_time), end_name, str(end_time), des, str(total_time)])
            except:
                pass

    df = pd.DataFrame(rows, columns=["route_id", "org", "srt_time", "srt_name", "bike_time", "end_name", "end_time", "des", "total_time"])
    df.set_index("route_id", inplace=True)
    return df.sort_values(by="total_time")
```

```python
read_routes('대한건축학회', '서울대학교 자하연')
```

```python
read_routes('서울시청', '서울고속터미널')
```
